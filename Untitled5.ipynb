```python
def baseline_pricing(prev_price, occupancy, capacity, alpha=0.05):
    """Simple linear model where price increases with occupancy."""
    occupancy_rate = occupancy / capacity
    new_price = prev_price + alpha * occupancy_rate * prev_price
    return np.clip(new_price, 5, 20)  # Bounded between $5 and $20
```

```python
def get_vehicle_weight(vehicle_type):
    return {'car': 1.0, 'bike': 0.5, 'truck': 1.5}.get(vehicle_type, 1.0)

def calculate_demand(row, weights):
    occ_rate = row['occupancy'] / row['capacity']
    vehicle_weight = get_vehicle_weight(row['vehicle_type'])
    demand = (
        weights['alpha'] * occ_rate +
        weights['beta'] * row['queue_length'] -
        weights['gamma'] * row['traffic'] +
        weights['delta'] * row['is_special_day'] +
        weights['epsilon'] * vehicle_weight
    )
    return demand

def demand_based_pricing(row, base_price=10, lambda_=0.4, weights=None):
    if weights is None:
        weights = {'alpha': 0.6, 'beta': 0.2, 'gamma': 0.1, 'delta': 0.5, 'epsilon': 0.3}

    demand = calculate_demand(row, weights)
    normalized_demand = (demand - row['demand_min']) / (row['demand_max'] - row['demand_min'] + 1e-5)
    price = base_price * (1 + lambda_ * normalized_demand)
    return np.clip(price, base_price * 0.5, base_price * 2)
```

```python
from geopy.distance import geodesic

def compute_nearest_competitor_prices(current_id, current_loc, all_data, radius_km=1):
    competitors = all_data[(all_data['id'] != current_id)]
    nearby = competitors[competitors.apply(
        lambda row: geodesic(current_loc, (row['latitude'], row['longitude'])).km < radius_km, axis=1)]

    if not nearby.empty:
        return nearby['price'].mean(), True
    else:
        return None, False

def competitive_pricing(row, all_data, base_price=10):
    avg_competitor_price, has_nearby = compute_nearest_competitor_prices(
        row['id'], (row['latitude'], row['longitude']), all_data)

    if has_nearby:
        if row['occupancy'] >= row['capacity']:
            return avg_competitor_price * 0.95  # reduce price to reroute
        elif avg_competitor_price > row['price']:
            return min(row['price'] * 1.1, avg_competitor_price)  # price hike
    return row['price']
```
